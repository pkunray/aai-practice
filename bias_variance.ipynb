```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def true_func(x):
    return 1.5*x - 0.2

def true_func2(x):
    return 1.1*x - 0.2

x = np.arange(0,1.00001,0.1)
y = np.asarray([true_func(xx) + np.random.randn()*0.15 for xx in x])

mod = np.poly1d(np.polyfit(x ,y , 11))
x_check = np.arange(0,1.00001,0.01)
y_trans = mod(x_check)

y_err = np.mean((y - mod(x))**2)

plt.title(f"MSE={y_err}")
plt.ylim([-1,3])
plt.plot(x,y, "ro")
plt.plot(x_check,y_trans)
```
