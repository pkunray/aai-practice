```python
import pandas as pd

# https://www.kaggle.com/datasets/rabieelkharoua/students-performance-dataset

df = pd.read_csv('/Users/kun/Desktop/SRH/Blocks/AAI/materials/Student_performance_data _.csv')
df.head(10)
```

```python
X = df.drop(columns=['StudentID', 'GPA','GradeClass'])
y = df['GradeClass']

# Whether should one-hot encode ParentalEducation? We can consider its steps.
X = pd.get_dummies(X,columns=['Gender','Ethnicity','ParentalEducation'], dtype=float)
features = X.columns
print(X.info())
X.head(10)
```

```python
# Standardize the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
print(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
X_pca = pca.fit_transform(X)
print(pca.explained_variance_)
#print(pca.explained_variance_ratio_)
```

```python
# PCA components DataFrame
pca_df = pd.DataFrame(pca.components_, columns=features)
pca_df
```
